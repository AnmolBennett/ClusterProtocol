```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

data = pd.read_csv('BTC-USD.csv')
data['Date'] = pd.to_datetime(data['Date'])

data.fillna(method='ffill', inplace=True)
data
```

```python
data['7-day MA'] = data['Close'].rolling(window=7).mean()
data['30-day MA'] = data['Close'].rolling(window=30).mean()

# Volatility
data['7-day Volatility'] = data['Close'].rolling(window=7).std()
data['30-day Volatility'] = data['Close'].rolling(window=30).std()
data
```

```python
def calculate_rsi(series, window):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data['RSI'] = calculate_rsi(data['Close'], window=14)
data
```

```python
data.dropna(inplace=True)
data
```

```python
scaler = MinMaxScaler()
numerical_features = ['Open', 'High', 'Low', 'Close', 'Volume', '7-day MA', '30-day MA', '7-day Volatility', '30-day Volatility', 'RSI']
data[numerical_features] = scaler.fit_transform(data[numerical_features])
data
```

```python
data.to_csv('preprocessed_data.csv', index=False)
print("Data preprocessing complete. Saved to 'data/preprocessed_data.csv'.")
```
